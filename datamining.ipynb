```python
import pandas as pd


sheets = pd.read_excel('教育科學.xlsx', sheet_name=None, engine='openpyxl')

df = pd.DataFrame()

sheet_names = list(sheets.keys())
sheet_names.reverse()

# 將所有工作表的資料合併到一個 DataFrame 中
for sheet_name in sheet_names[:-1]:  # 不包括最後一張表
    df = sheets[sheet_name]
    df = pd.concat([df, df], ignore_index=True)

print(df.columns)
df
```

```python
# 刪除 '名次' 欄位
df = df.drop(columns=['名次'])

# 檢查結果
df
```

```python
# 計算每個提名單位在同一個職類名稱中出現的次數
df['得名次數'] = df.groupby(['職類名稱', '提名單位'])['提名單位'].transform('count')

# 刪除重複的行，只保留第一個出現的
df = df.drop_duplicates(subset=['職類名稱', '提名單位'])

print(df.head(20))
# unique_units = df['提名單位'].unique()

# 列印結果
# for unit in unique_units:
#     print(unit)
# df.loc[df['學校名稱'] == '國立臺灣大學', '是否得名'] = '0'
```
